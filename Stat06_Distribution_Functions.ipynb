# Normal Distribution Function

```r
rnorm(5) # default: mean = 0, sd = 1
```

```r
rnorm(5, mean = 100, sd = 5)
```

```r
x <- seq(0, 100, 10)
y <- rnorm(x, mean = 50, sd = 5)
plot(x,y,
    ylim = c(0, max(y))
    ) 
```

# Probability Density Function

```r
x <- seq(-4, 4, 0.1)
y <- dnorm(x, mean = 0, sd = 1)
plot(x,y, 
    type = "l",
    )
```

### Comparison  to histogram

```r
hist(iris$Sepal.Width,
    col = "mistyrose",
     main = "iris Sepal.Width histogram and probability density",
    xlim = c(2.0, 4.5),
     freq = F
    )

lines(density(iris$Sepal.Width),
      col = "red")

x <- seq(2.0, 4.5, 0.1)
curve(dnorm(x, mean = mean(iris$Sepal.Width), sd = sd(iris$Sepal.Width)),
      col = "blue",
      add = T)
```

# Cumulative Distribution Function

```r
x <- seq(-4, 4, 0.1)
y <- pnorm(x, mean = 0, sd = 1)
plot(x,y,                       # to check if data is asymmetric or not
    type = "l")
```

# Math Function

```r
f <- function(x) {
    y = x^2
    return(y)
}

x <- seq(-5, 5, 0.1)
plot(x, f(x), type = "l")
```

```r
f <- function(x) {
    y = exp(-x) - 4 * x^2 + x + 3 * abs(x) - 3  # e^-x = exp(-x)
    return(y)
}

x <- seq(-5, 5, 0.1)
plot(x, f(x), type = "l")
```
