```r
TSdata <- read.csv("./Data/MyData.csv")
summary(TSdata)
```

# plot

```r
plot(TSdata$date, TSdata$SmallGrowth, # x, y,
     type = "b",
     lty = 3,
     pch = 23,
     col = "red",
     cex = 1.3,
     main = "Small Growth",
     sub = "1979 - 2011",
     xlab = "date",
     ylab = "growth",
     xlim = c(197901, 201104), 
     ylim = c(-35, 60),
     lwd = 1.5,
     bg = "orange",
#     axes = F
#     ann = F
    )
```

```r
# Alternatively,
plot(TSdata$SmallGrowth ~ TSdata$date, # y ~ x,
     type = "b",
     lty = 3,
     pch = 23,
     col = "red",
     cex = 1.3,
     main = "Small Growth",
     sub = "1979 - 2011",
     xlab = "date",
     ylab = "growth",
     xlim = c(197901, 201104), 
     ylim = c(-35, 60),
     lwd = 1.5,
     bg = "orange",
#     axes = F
#     ann = F
    )
```

```r
# cf. reverse,
plot(TSdata$date, TSdata$SmallGrowth, 
     type = "b",
     lty = 3,
     pch = 23,
     col = "dark blue",     
     cex = 1.3,
     main = "Small Growth",
     sub = "1979 - 2011",
     xlab = "date",
     ylab = "growth",
     xlim = c(201104, 197901), # exchange min <> max
     ylim = c(-35, 60),
     lwd = 1.5,
     bg = "blue"
#     ann = F
    )
```

# Separate style options

```r
plot(TSdata$date, TSdata$SmallGrowth, # x, y,
     type = "b",
     col = "red",
     axes = F,
     ann = F
    )

axis(1, xlim = TSdata$date)

axis(2, ylim = c(0, 60))
     
title(main="Small Growth",col.main="purple")

title(xlab = "date", col.lab="green")

title(ylab = "growth", col.lab="blue")



#      lty = 3,
#      pch = 23,
#      cex = 1.3,
#      main = "Small Growth",
#      sub = "1979 - 2011",
#      xlab = "date",
#      ylab = "growth",
#      xlim = c(197901, 201104), 
#      ylim = c(-35, 60),
#      lwd = 1.5,
#      bg = "orange",

#     )
```

#  Multiple plots in one page

```r
par(mfrow = c(2,1))
plot(TSdata$SmallGrowth ~ TSdata$date, # y ~ x,
     type = "b",
     lty = 3,
     pch = "o",
     col = "red",
     main = "Small Growth",
     sub = "1979 - 2011",
     xlab = "date",
     ylab = "growth",
     xlim = c(197901, 201104), 
     ylim = c(-35, 60),
#     ann = F
    )

# cf. reverse,
plot(TSdata$date, TSdata$SmallGrowth, 
     type = "b",
     lty = 3,
     pch = "o",
     col = "blue",
     main = "Small Growth",
     sub = "2011 - 1979",
     xlab = "date",
     ylab = "growth",
     xlim = c(201104, 197901), # exchange min <> max
     ylim = c(-35, 60),
#     ann = F
    )
```

# Add lines

```r
baseball   <- c(110, 300, 150, 280, 310)
basketball <- c(180, 200, 210, 190, 170)
tennisball <- c(210, 150, 160, 210,  70)
```

```r
plot(baseball,
    type = "o",
    col = "red",
    lty = 2,
    pch = 24,
    cex = 1.5,
    main = "Sales Figures",
    sub = "2019",
    xlab = "Month",
    ylab = "Sales",
    xlim = c(1, 5), 
    ylim = c(0, 350),
    lwd = 1.5,
    bg = "orange",
#     axes = F
     )

lines(basketball,
      type = "o",
      pch = 21,
      col = "dark green",
      lty = 2) 

lines(tennisball,
      type = "b",
      pch = 22,
      col = "blue",
      lty = 2)

legend(1, 75, 
       c("BaseBall", "SoccerBall", "BeachBall"), 
       cex = 1, 
       col = c("Red", "Dark Green", "Blue"), 
       pch = 21, 
       lty = 1: 3)
```
