```r
TSdata <- read.csv("./Data/MyData.csv")
summary(TSdata)
```

# plot

```r
plot(TSdata$date, TSdata$SmallGrowth, # x, y,
     type = "b",
     lty = 3,
     pch = "o",
     col = "red",
     main = "Small Growth",
     sub = "1979 - 2011",
     xlab = "date",
     ylab = "growth",
     xlim = c(197901, 201104), 
     ylim = c(-35, 60),
#     ann = F
    )
```

```r
# Alternatively,
plot(TSdata$SmallGrowth ~ TSdata$date, # y ~ x,
     type = "b",
     lty = 3,
     pch = "o",
     col = "red",
     main = "Small Growth",
     sub = "1979 - 2011",
     xlab = "date",
     ylab = "growth",
     xlim = c(197901, 201104), 
     ylim = c(-35, 60),
#     ann = F
    )
```

```r
# cf. reverse,
plot(TSdata$date, TSdata$SmallGrowth, 
     type = "b",
     lty = 3,
     pch = "o",
     col = "blue",
     main = "Small Growth",
     sub = "2011 - 1979",
     xlab = "date",
     ylab = "growth",
     xlim = c(201104, 197901), # exchange min <> max
     ylim = c(-35, 60),
#     ann = F
    )
```

#  Multiple plots in one page

```r
par(mfrow = c(2,1))
plot(TSdata$SmallGrowth ~ TSdata$date, # y ~ x,
     type = "b",
     lty = 3,
     pch = "o",
     col = "red",
     main = "Small Growth",
     sub = "1979 - 2011",
     xlab = "date",
     ylab = "growth",
     xlim = c(197901, 201104), 
     ylim = c(-35, 60),
#     ann = F
    )

# cf. reverse,
plot(TSdata$date, TSdata$SmallGrowth, 
     type = "b",
     lty = 3,
     pch = "o",
     col = "blue",
     main = "Small Growth",
     sub = "2011 - 1979",
     xlab = "date",
     ylab = "growth",
     xlim = c(201104, 197901), # exchange min <> max
     ylim = c(-35, 60),
#     ann = F
    )
```

```r
par(mfrow = c(1,2))
plot(TSdata$SmallGrowth ~ TSdata$date, # y ~ x,
     type = "b",
     lty = 3,
     pch = "o",
     col = "red",
     main = "Small Growth",
     sub = "1979 - 2011",
     xlab = "date",
     ylab = "growth",
     xlim = c(197901, 201104), 
     ylim = c(-35, 60),
#     ann = F
    )

# cf. reverse,
plot(TSdata$date, TSdata$SmallGrowth, 
     type = "b",
     lty = 3,
     pch = "o",
     col = "blue",
     main = "Small Growth",
     sub = "2011 - 1979",
     xlab = "date",
     ylab = "growth",
     xlim = c(201104, 197901), # exchange min <> max
     ylim = c(-35, 60),
#     ann = F
    )
```
